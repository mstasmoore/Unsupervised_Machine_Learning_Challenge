```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
```

```python
crypto_df = pd.read_csv('Resources/crypto_data.csv')
crypto_df
```

```python
#non traded crypto 

crypto_df = crypto_df[crypto_df.IsTrading == True]
crypto_df
```

```python
#remove IsTrading column 

crypto_df = crypto_df.drop(columns='IsTrading')
crypto_df
```

```python
# Remove NAN

crypto_df.dropna(inplace=True)
crypto_df
```

```python
# Filter mined to be greater than zero.

crypto_df = crypto_df[crypto_df["TotalCoinsMined"]> 0]
crypto_df
```

```python
# deleting non numeric coloum coinname

crypto_df = crypto_df.drop(columns='CoinName', axis=1)
crypto_df
```

```python
crypto_df['ProofType'].nunique()
```

```python
crypto_df['Algorithm'].nunique()
```

```python
#index names

crypto_df.index.names = ['']
crypto_df
```

```python
#convert the remaining values with text values into numebers. 

crypto_numeric_df = pd.get_dummies(crypto_df, columns=['Algorithm', 'ProofType'])
crypto_numeric_df.head(50)
```

```python
#set data so columns that contain larger values dont change outcome.
X_scaled = StandardScaler().fit_transform(crypto_numeric_df[['TotalCoinsMined','TotalCoinSupply']])
X_scaled
```

```python
X_scaled_df = crypto_numeric_df.copy()
X_scaled_df['TotalCoinsMined'] = X_scaled[:,0]
X_scaled_df['TotalCoinSupply'] = X_scaled[:,1]
X_scaled_df
```

```python
#dimensionality reduction with

pca = PCA(n_components= 0.90)
crypto_pca  = pca.fit_transform(X_scaled_df)
crypto_pca.shape
```

```python
#add to data and create new dataframe
crypto_pca_df = pd.DataFrame(
    data=crypto_pca, index=crypto_numeric_df.index
)
crypto_pca_df
```

```python
#use tnse model

tsne = TSNE(learning_rate=35)
```

```python
#trying to run model
tsne_features = tsne.fit_transform(crypto_pca_df)
tsne_features
```

```python
#scatter plot 
tsne_df = pd.DataFrame(tsne_features, columns=['x','y'])
tsne_df
```

```python
# plot
plt.scatter(tsne_df['x'], tsne_df['y'],color="#AC3B61")
plt.show()
```

```python
#elbow plot to test for the best number of clusters.

inertia = []
k = list(range(1, 11))

for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(tsne_df)
    inertia.append(km.inertia_)

#hvPlot
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow
```

```python
#plot to find k
plt.plot(df_elbow['k'], df_elbow['inertia'],color="#BAB2B5")
plt.xticks(range(1,11))
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow curve')
plt.show()
```
